```python
# read the property files

# read the data with properties
folder_path_props = "./datasets/PropsQM9/"

# Get all files from the directory
all_files = os.listdir(folder_path_props)

# glob.glob returns a list of file paths in the specified directory that match the pattern '*.xyz'
xyz_files = glob.glob(os.path.join(folder_path_props, '*.xyz'))
```

```python
# read in the file structures
def read_xyz_file(filepath):
    with open(filepath, 'r') as f:
        lines = f.readlines()

    num_atoms = int(lines[0].strip())
    properties_line = lines[1].split('\t')
    identifier = properties_line[0]
    properties = list(map(float, ' '.join(properties_line[1:]).split()))  # join all elements after the first, then split by space and convert to float
    atoms = [line.split() for line in lines[2:2+num_atoms]]
    extra_data = [line.strip() for line in lines[2+num_atoms:] if line.strip()]

    atoms_df = pd.DataFrame(atoms, columns=["Atom", "X", "Y", "Z", "Extra_Value"])

    return identifier, atoms_df, properties, extra_data
```

```python
list_of_dataframes = []
for file in xyz_files:
    try:
        identifier, atoms_df, properties, extra_data = read_xyz_file(file)
        list_of_dataframes.append([identifier, atoms_df, properties, extra_data])
    except ValueError as ve:
        print(f"Skipping file {file} due to error: {ve}")
```

```python
list_of_properties = []
for i in range(len(list_of_dataframes)):
    DipoleIn = list_of_dataframes[i][2][3]
    GapIn = list_of_dataframes[i][2][7]
    SMILESCodeIn =list_of_dataframes[i][3][1].split('\t')[0]
    #SMILEtest =list_of_dataframes[i][3][1]
    #SMILESCodeIn = SMILEtest.split()[0]
    list_of_properties.append([SMILESCodeIn, GapIn, DipoleIn])
```
